```python
import seaborn as sns
import pandas as pd
```

```python
tips = pd.read_csv('seaborn-data/tips.csv')
```

```python
tips.head()
```

Above dataset is of tips dataset which contains how many people had a dinner and how much tip they paid

Distplot will gives us univariate set of observations in dataset of one variable

```python
sns.distplot(tips['total_bill'])
```

In above plot we get a histogram and kde i.e kernel distribution estimation, if we want to remove that argument we can remove it by saying 'kde=False' like below

```python
sns.distplot(tips['total_bill'],kde=False)
```

we can give bins as a argument in distplot to see detail variation of total bill

```python
sns.distplot(tips['total_bill'],kde=False,bins=40)
```

Joint plot is used to bivariate set i.e more than one value like in our case we can check what is tip and total bill below using joint plot

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

jointplot has also one value called 'kind' which gives you what is happening in the data more like below

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

pairplot plots pairwise relationships across an entire data frame at least for the numerical coloumns, it also plots pairwise relationship across entire dataset

```python
sns.pairplot(tips)
```

we can add hue argument in the pairplot, hue argument will take coloumn name if categorical coloumn and categorical does not mean numerical or non numerical, it means the coloumn which has categories like 'sex' colomn is categorical in our dataset

```python
sns.pairplot(tips,hue='sex')
```
